```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
# если слишком много данных, то verbose поможет. Посчитать сколько пропущенных null_counts.
train.info(verbose=True, memory_usage=True, null_counts=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
from ipywidgets import Image
from io import StringIO
import pydotplus
from sklearn.tree import export_graphviz

dot_data = StringIO()
export_graphviz(dtc, feature_names=train.columns, out_file=dot_data, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(value=graph.create_png())
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
import math
```

```python
# фунция демонстрирующая что можно из python взять коэффиценты и посчтитать вероятность.
def predict(featchers, coef, last_coef):
    s = 0
    for x1, x2 in zip(featchers, coef):
        s += x1*x2
    s += last_coef
    return   (
      1/(
          1 + math.e**(
            -1 * (
                s
            )
          )
      )
  )
```

```python
predict([1,1], [1,1], 1)
```

```python
predict(
list(X_test.iloc[0]), lr.coef_[0], lr.intercept_[0]
    )
```
